```python
import numpy as np
def pr(n):
  p = 2 * 2**0.5
  k = 0.0
  r = 2.0
  d = 100.0
  for i in range(2, n, 1):
    d = (p - np.pi) ** 2
    p = float(2**i * (2 * (1 - (1 - p ** 2 / 2**(2 * i))**0.5))**0.5)
    if (p - np.pi) ** 2 < d:
        k = p
    r = r / (2 + (4 - r)**0.5)
  print("n =", n)
  print("P", k)
  print("R:", 2 ** (n - 1) * r ** 0.5, "\n")
for n in range(3, 61):
  pr(n)
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
n = 60
X = []
Y = []
for i in range(n):
    X.append(float(i) / n)
    Y.append((float(i) / n) ** 0.5)

x_ = 0
y_ = 0
for i in range(n):
    x_ = x_ + X[i]
    y_ = y_ + Y[i]
x_ = x_ / n
y_ = y_ / n

r = 0
s = 0
for i in range(n):
    r = r + (X[i] - x_) * (Y[i] - y_)
    s = s + (X[i] - x_) ** 2
a = r / s
b = y_ - a * x_
print(a, b)


plt.plot(np.array(X), np.array(Y))
plt.plot(np.array(X), a * np.array(X) + b)
plt.show()
```
